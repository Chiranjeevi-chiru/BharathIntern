```python
x=data.drop(columns=['price'])
y=data['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
column_trans = make_column_transformer((OneHotEncoder(), ['beds']),remainder='passthrough')
```

```python
scalar=StandardScaler()
```

```python
lr=LinearRegression()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
x_scaled=scalar.fit_transform(x)
lr=LinearRegression()
lr.fit(x_scaled,y)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression

pipe = make_pipeline(StandardScaler(with_mean=False), LinearRegression())

pipe.fit(x_train, y_train)

y_pred_lr = pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred_lr)
```

```python
lasso=Lasso()
```

```python
pipe=make_pipeline(column_trans,scalar,lasso)
```

```python
pipeline = make_pipeline(LinearRegression())
pipeline.fit(x_train, y_train)
y_pred_lasso = pipeline.predict(x_test)
```

```python
pipe = make_pipeline(StandardScaler(with_mean=False, with_std=False), LinearRegression())
pipe.fit(x_train, y_train)
y_pred = pipe.predict(x_test)
r2_score(y_test,y_pred)
```

```python
pipe = make_pipeline(column_trans, Ridge())
pipe.fit(x_train, y_train)
y_pred_ridge = pipe.predict(x_test)
r2_score(y_test, y_pred_ridge)
```

```python
print("No Regularization: ", r2_score(y_test,y_pred_lr))
print("Lasso:", r2_score(y_test,y_pred))
print("Ridge:",r2_score(y_test,y_pred_ridge))
```
